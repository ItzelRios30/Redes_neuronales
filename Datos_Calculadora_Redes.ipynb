```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=40)
```

```python
## Normalizar los datos
media = X_train.mean(axis=0)
X_train -= media
std = X_train.std(axis=0)
X_train /= std

X_test -= media
X_test /= std

mediay = y_train.mean(axis = 0)
stdy = y_train.std(axis = 0)

y_train -= mediay
y_train /= stdy

y_test -= mediay
y_test /= stdy
```

```python
from keras import models
from keras import layers
from keras import optimizers
```

```python
def build_model():
    model = models.Sequential()
    model.add(layers.Dense(128, activation = 'relu',
                          input_shape = (X_train.shape[1],)))
    model.add(layers.Dense(120, activation = 'relu'))
    model.add(layers.Dense(112, activation = 'relu'))
    model.add(layers.Dense(104, activation = 'relu'))
    model.add(layers.Dense(96, activation = 'relu'))
    model.add(layers.Dense(88, activation = 'relu'))
    model.add(layers.Dense(80, activation = 'relu'))
    model.add(layers.Dense(72, activation = 'relu'))
    model.add(layers.Dense(64, activation = 'relu'))
    model.add(layers.Dense(56, activation = 'relu'))
    model.add(layers.Dense(48, activation = 'relu'))
    model.add(layers.Dense(40, activation = 'relu'))
    model.add(layers.Dense(32, activation = 'relu'))
    model.add(layers.Dense(24, activation = 'relu'))
    model.add(layers.Dense(16, activation = 'relu'))
    model.add(layers.Dense(8, activation = 'relu'))
    model.add(layers.Dense(1))
    model.compile(optimizer = optimizers.RMSprop(lr=0.001),
                  loss = 'mse', metrics =  ['mae'])
    return model
```

```python
model = build_model()
history = model.fit(X_train, y_train, epochs=25000, batch_size=309)
```

```python
results = model.evaluate(X_test, y_test)
```

```python
import matplotlib.pyplot as plt
history_dict = history.history
loss_values = history_dict['loss']
ErroMedioAbsoluto = history_dict['mae']
epochs = range(1, 25000 + 1)
plt.plot(epochs, loss_values, 'b', label='Training loss')
plt.plot(epochs, ErroMedioAbsoluto, 'k', label='Training mae')
plt.title('Training and MAE loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
n = 14
m = 2433 


X_tensor = torch.from_numpy(X_train.values.reshape(m,n))
y_tensor = torch.from_numpy(y_train.values.reshape(m,1))

print(X_train.shape)
print(y_train.shape)
print(X_tensor.shape)
print(y_tensor.shape)
```

```python
A = torch.randn((1, n), requires_grad=True)
b = torch.randn(1, requires_grad=True)

# modelo predictivo
def model(x_input):
    return A.mm(x_input) + b
###función de perdida ###
def loss(y_predicted, y_target):
    return ((y_predicted - y_target)**2).sum()
```

```python
optimizer = optim.Adam([A, b], lr=0.01)
#optimizer = optim.ASGD([A, b], lr=0.01)
# loop de optimizacion 
for t in range(1000):
    #  gradiente =  0.
    optimizer.zero_grad()
    # Calcular los valores precteivos de Y de la tabla train 
    y_predicted = model(X_tensor.t().float())
    # error de predicción
    current_loss = loss(y_predicted, y_tensor)
    # Calcular el gradiente de perdida con resepcto a A y B  
    current_loss.backward()
    # Actualizar A y b 
    optimizer.step()
print(f"t = {t}, loss = {current_loss}, A = {A.detach().numpy()}, b = {b.item()}")
```
